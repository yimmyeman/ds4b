```python
# Variables Categóricas

X_cat = employee_df.describe(include = ['O']).columns.tolist()
X_cat = employee_df[X_cat]
X_cat
```

```python
X_cat2 = X_cat
```

```python
# Variables Categóricas Forma 1

from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
X_cat
```

```python
# Variables Categóricas Forma Alternativa

def createDummies(df, var_name):
    '''
    Función para convertir variables categóricas en variables Dummies.
    La función borra var_name una vez ha sido transformada.
    
    '''
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy ], axis = 1)
    return df

list_cat =  employee_df.describe(include = ['O']).columns.tolist()

# Se crea un bucle for para recorrer las variables y transformarlas
for var in list_cat:
    X_cat2 = createDummies(X_cat2, var)
    
X_cat2
```

```python
X_cat2.shape
```

```python
temp_list = employee_df.describe().columns.tolist()
del temp_list[1] # Eliminando la variable dependiente 'Atrittion'
temp_list
```

```python
# Notemos que hemos elminado la columna target, 'Atrittion'
X_numerical = employee_df[temp_list]
X_numerical
```

```python
X_all = pd.concat([X_cat2, X_numerical], axis = 1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

# TAREA #7: ENTRENAR Y EVALUAR UN CLASIFICADOR UTILIZANDO LA REGRESIÓN LOGÍSTICA

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```


\

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format(100*accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
```

El número de falsos positivos es bastante algo como para considerarlo un buen modelo.

```python
print(classification_report(y_test, y_pred))
```

# TAREA #9: ENTRENAR Y EVALUR UN CLASIFICADOR UTILIZANDO DEEP LEARNING

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu', input_shape = (50,)))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = (y_pred > 0.5)
y_pred
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title("Función de pérdida del modelo durante la fase de Entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Error de Entrenamiento")
plt.legend(["Error de Entrenamiento"])
```

```python
plt.plot(epochs_hist.history['accuracy'])
plt.title("Función de pérdida del modelo durante la fase de Entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Accuracy de Entrenamiento")
plt.legend(["Accuracy de Entrenamiento"])
```
